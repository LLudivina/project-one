```python
import pandas as pd
from datetime import datetime
```

```python
# 2014
# Read the data into a Pandas DataFrame
df = pd.read_csv("climate_data2014.csv")
df.head(10)
```

```python
df.columns
```

```python
df.tail()
```

```python
df[['WND_direction', 'WND_dQC', 'WND_type', 'WND_speed', 'WND_speedQC']] = df['WND'].str.split(',', expand=True)
df.head()
```

```python
df[['SKY_ceiling', 'SKY_ceilingQC', 'SKY_ceilingDC', 'SKY_CAVOK']] = df['CIG'].str.split(',', expand=True)
df.head()
```

```python
df[['VIS_distance', 'VIS_distanceQC', 'VIS_variability', 'VIS_variabilityQC']] = df['VIS'].str.split(',', expand=True)
df.head()
```

```python
df[['AIR_temp', 'AIR_tempQC']] = df['TMP'].str.split(',', expand=True)
df.head()
```

```python
df[['DEW_temp', 'DEW_tempQC']] = df['DEW'].str.split(',', expand=True)
df.head()
```

```python
df[['SLP_pressure', 'SLP_pressureQC']] = df['DEW'].str.split(',', expand=True)
df.head()
```

```python
clean_df = df.drop(columns=['WND', 'CIG', 'VIS', 'TMP', 'DEW', 'SLP'])
clean_df.head()
```

```python
clean_df.tail()
```

```python
# Dropping the null rows
df1 = clean_df[clean_df["CALL_SIGN"].str.contains("9999") == False]
df2 = df1[df1["WND_direction"].str.contains("999") == False]
df3 = df2[df2["WND_dQC"].str.contains("9") == False]
df4 = df3[df3["WND_type"].str.contains("9") == False]
df5 = df4[df4["WND_speed"].str.contains("9999") == False]
df6 = df5[df5["WND_speedQC"].str.contains("9") == False]
df7 = df6[df6["SKY_ceiling"].str.contains("99999") == False]
df8 = df7[df7["SKY_ceilingQC"].str.contains("9") == False]
df9 = df8[df8["SKY_ceilingDC"].str.contains("9") == False]
df10 = df9[df9["SKY_CAVOK"].str.contains("9") == False]
df11 = df10[df10["VIS_distance"].str.contains("999999") == False]
df12 = df11[df11["VIS_distanceQC"].str.contains("9") == False]
df13 = df12[df12["VIS_variability"].str.contains("9") == False]
df14 = df13[df13["VIS_variabilityQC"].str.contains("9") == False]
df15 = df14[df14["AIR_temp"].str.contains("9999") == False]
df16 = df15[df15["AIR_tempQC"].str.contains("9") == False]
df17 = df16[df16["DEW_temp"].str.contains("9999") == False]
df18 = df17[df17["DEW_tempQC"].str.contains("9") == False]
df19 = df18[df18["SLP_pressure"].str.contains("9999") == False]
df20 = df19[df19["SLP_pressureQC"].str.contains("9") == False]

df20.tail()
```
